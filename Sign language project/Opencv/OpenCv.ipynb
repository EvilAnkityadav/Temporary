# Working with Image

```python
import cv2
import os
```

```python
img_path=os.path.join('data','images','image1.jpg')
```

```python
img_path
```

```python
img = cv2.imread(img_path)
```

```python
img.shape
```

```python
from matplotlib import pyplot as plt
```

```python
plt.imshow(img)
```

```python
recolor=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(grey)
plt.show()
plt.imshow(recolor)
plt.show()
```

```python
output_path=os.path.join('data','images','gray.jpg')
cv2.imwrite(output_path,grey)
```

# Working with videos

```python
cap=cv2.VideoCapture(os.path.join('data','videos','video1.mp4'))
for frame_idx in range (int(cap.get(cv2.CAP_PROP_FRAME_COUNT))): 
    _,frame=cap.read()
    gray_video=cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    cv2.imshow('Video Player',gray_video)
    if(cv2.waitKey(10) & 0xFF==ord('q')):
        break
cap.release()
cv2.destroyAllWindows()
```
